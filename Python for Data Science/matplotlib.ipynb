```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.xlabel('X Axis of subplot 1')
plt.ylabel('Y Axis of subplot 1')
plt.title('Title of Subplot 1')
plt.plot(x,y,'r*-')
plt.xlabel('X Axis of subplot 1')
plt.ylabel('Y Axis of subplot 1')
plt.title('Title of Subplot 1')

plt.subplot(1,3,3)
plt.plot(x,y,'m--')
plt.xlabel('X Axis of subplot 2')
plt.ylabel('Y Axis of subplot 2')
plt.title('Title of Subplot 2')
```

```python
#OO 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'*-')
axes1.set_xlabel('X Axis of Larger Plot')
axes1.set_ylabel('Y Axis of Larger Plot')
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_xlabel('X Axis of Smaller Plot')
axes2.set_ylabel('Y Axis of Smaller Plot')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X Axis of Larger Plot')
axes[0].set_ylabel('Y Axis of Larger Plot')
axes[0].set_title('Larger Plot')
axes[1].plot(y,x,'y--')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y,'m*-')
axes[1].plot(y,x,'g.-')
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=100)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')
axes.legend(loc=(0.5,0.5))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',linewidth=3,alpha=0.5,linestyle='-',marker='*',markersize=10,markerfacecolor='purple',markeredgewidth=1.5,markeredgecolor='blue')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,'r--')
axes.set_xlim([0,1])
axes.set_ylim([0,1])
```

## advanced matplotlib concepts from here onwards

```python
fig,axes  = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
axes[0].plot(x,x**2)
axes[0].plot(x,np.exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title("Logarithmic Scale(y)")
```

```python
fig,axes  = plt.subplots(nrows=1,ncols=2,figsize=(10,6))
axes[0].plot(x,x**2)
axes[0].plot(x,np.exp(x))
axes[0].set_title("Normal Scale")

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale('log')
axes[1].set_xscale('log')
axes[1].set_title("Logarithmic Scale")
```
