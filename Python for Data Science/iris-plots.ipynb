```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris = pd.read_csv('C:\\Users\\User\\python for data science\\datasets\\IRIS.csv')
```

```python
iris.head()
```

```python
sns.set_style(style='darkgrid')
```

```python
sns.barplot(x='sepal_length',y='sepal_width',data=iris)
plt.tight_layout()
```

```python
sns.countplot(x='species',data=iris,palette='Pastel2')
```

```python
sns.pairplot(data=iris,hue='species',palette='RdPu')
```

```python
sns.distplot(iris['sepal_length'],color='brown',bins=50)
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g  = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_lower(plt.scatter)
g.map_upper(sns.distplot,hist=False)
```

```python
iris.plot.area(stacked=False,colormap='coolwarm')
```

```python
iris.plot.line(stacked=False,colormap='coolwarm')
```

```python
iris[['sepal_width','sepal_length']].plot.area(stacked=False,colormap='copper',alpha=0.4)
```
