```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
titan = sns.load_dataset('titanic')
```

```python
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titan.head()
```

```python
g = sns.JointGrid(x='fare',y='age',data=titan)
g.plot_joint(plt.scatter)
g.plot_marginals(sns.distplot, kde=False)
```

```python
sns.jointplot(x='fare',y='age',data=titan)
```

```python
sns.distplot(titan['fare'],kde=False,bins=30,color='red')
```

```python
sns.boxplot(x='class',y='age',data=titan,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titan,palette='magma')
```

```python
sns.countplot(x='sex',data=titan,palette='coolwarm')
```

```python
titan.corr()
```

```python
sns.heatmap(titan.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titan,col='sex')
g = g.map(plt.hist,'age')
```
