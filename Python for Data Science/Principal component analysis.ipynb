```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer_df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
cancer_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(cancer_df)
```

```python
scaled_cancer_df = scaler.transform(cancer_df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_cancer_df)
```

```python
x_pca = pca.transform(scaled_cancer_df)
```

```python
x_pca.shape
```

```python
scaled_cancer_df.shape
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
```

```python
pca.components_
```

```python
cancer_df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
cancer_df_comp.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(cancer_df_comp)
```
